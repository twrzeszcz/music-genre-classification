## Music genre classification project

In this project we will try to classify music genres with the dataset from [kaggle](https://www.kaggle.com/andradaolteanu/gtzan-dataset-music-genre-classification). It consists of 100 audio files for each of 10 music genres. Audio file is a 30 second clip. To classify music genres one can utilize CNNs for which an audio file has to be represented as an image. The most efficient representation is the so called *mel spectrogram* or a heatmap which shows amplitude of each frequency for every time step. Model is deployed on [Streamlit](https://share.streamlit.io/twrzeszcz/music-generation-streamlit/main/app_streamlit_merged.py) where user can upload the audio file which is then preprocessed and the genre is predicted.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
from tensorflow import keras
import librosa
import librosa.display
import cv2
import glob
import os
from tqdm.notebook import tqdm
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import io

%matplotlib inline
```

### Preprocessing

Spectrogram images are already provided for each audio file but they contain white bars which should not be used for model training. Therefore we will generate spectrograms directly from audio files using *librosa* library. Additionally we can split the 30 seconds audio clips into 10 parts to increase the dataset size. The raw dataset contains 1000 audio files which will give only 1000 spectrograms. If we split them then there will be 1000 spectrograms for each class. I have tested 2 models where the first one was trained on the 30s clips and the second one on 3s clips.

Lets first see how the spectrogram looks like. We have to load an audio file using *librosa* which will give us sampled time series and a sampling rate. Then we plug them into *melspectrogram* function which performs Fourier Transform and gives the power spectrum. After converting power spectrum to dB, we can call the *display.specshow* function which displays the spectrogram.

```python
class_names = [i.split('\\')[1] for i in glob.glob('Data/images_original/*')]
```

```python
class_names
```

```python
y, sr = librosa.load('Data\\genres_original\\blues\\blues.00010.wav')
S = librosa.feature.melspectrogram(y=y, sr=sr)
S_dB = librosa.power_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, ax=ax)
# ax.axis('off')
fig.colorbar(img, ax=ax, format='%+2.0f dB')
# plt.savefig('sample.png', 
#             bbox_inches='tight', pad_inches = 0.0, dpi=170)
```

Now we have to write a preprocessing function which will iterate over directiories with audio files and extract spectrograms which have to be saved in the separate location as images. Spectrograms are generated in the same way as it was described before. The only thing that we need to take care off is to hide axes and remove white bars so that the saved image only contains the spectrogram. This iterative process consumes lots of memory so one can generate spectrograms for only part of classes and after reloading notebook finish the rest. I had to do this for 3s clips where we have 10 spectrograms for each audio file. In *prep_audio_modified* function we have to additionally split audio files into 3s clips. This is done sequentially and we have to also iterate over every single clip within a track. 

### Preprocess full 30s clips

```python
def prep_audio():
    for i, k in zip(glob.glob('Data/genres_original/*'), glob.glob('data_prep*')):
        #get times series and sampling rate for each sample
        ts_sr = [librosa.load(j) for j in glob.glob(os.path.join(i, '*.wav'))]
        sr = [j[1] for j in ts_sr]
        spectrograms = [librosa.power_to_db(librosa.feature.melspectrogram(y=j[0], sr=j[1]), ref=np.max) for j in ts_sr]
        
        for idx, spec_sr in enumerate(zip(tqdm(spectrograms), tqdm(sr))):
            fig, ax = plt.subplots(figsize=(15, 5))
            img = librosa.display.specshow(spec_sr[0], x_axis='time', y_axis='mel', sr=spec_sr[1], ax=ax)
            ax.axis('off')
            plt.savefig(k + '\\' + k.split('\\')[1] + str(idx) + '.png', 
                        bbox_inches='tight', pad_inches = 0.0)
            plt.close()
```

```python
prep_audio()
```

### Preprocess 3s clips

```python
sr = 22050
def prep_audio_modified(sr):
    for i, p in zip(glob.glob('Data/genres_original/*')[0:4], glob.glob('data_prep_modified/*')[0:4]):
        ts_sr = [[librosa.load(j, offset=k, duration=3) for k in np.arange(0, 30, 3)] for j in glob.glob(os.path.join(i, '*.wav'))]
        spectrograms = [[librosa.power_to_db(librosa.feature.melspectrogram(y=j[0], sr=sr), ref=np.max) for j in element] for element in ts_sr]
    
        for idx_0, j in enumerate(spectrograms):
            for idx_1, spec in enumerate(tqdm(j)):
                fig, ax = plt.subplots(figsize=(15, 5))
                img = librosa.display.specshow(spec, x_axis='time', y_axis='mel', sr=sr, ax=ax)
                ax.axis('off')
                plt.savefig(p + '\\' + p.split('\\')[1] + str(idx_0) + str(idx_1) + '.png', 
                            bbox_inches='tight', pad_inches = 0.0)
                plt.close()
```

### Training

To train the model we can use the *flow_from_directory* function which generates data from specified directories one batch at a time during training so that the entire dataset is not loaded into memory. Size of images is chosen according to the requirements of pretrained model which will be used. All images are shuffled and splitted into train and validation sets with 80/20 ratio. Batch size is set to 16. Images are normalized already in the model architecture (*EfficientNet*) so no additional preprocessing is needed. That is why we only specify the validation set size in the *ImageDataGenerator*.  

### 30s

```python
datagen = keras.preprocessing.image.ImageDataGenerator(validation_split=0.2)

img_size = (224, 224)
path_to_data = 'data_prep'
train_generator_full = datagen.flow_from_directory(path_to_data,
                                              target_size=img_size,
                                              batch_size=16,
                                              seed=4000,
                                              shuffle=True,
                                              class_mode='sparse',
                                              subset='training')

valid_generator_full = datagen.flow_from_directory(path_to_data,
                                              target_size=img_size,
                                              batch_size=16,
                                              seed=4000,
                                              shuffle=False,
                                              class_mode='sparse',
                                              subset='validation')
```

### 3s

```python
datagen = keras.preprocessing.image.ImageDataGenerator(validation_split=0.2)

path_to_data = 'data_prep_modified'
train_generator_trimmed = datagen.flow_from_directory(path_to_data,
                                              target_size=img_size,
                                              batch_size=16,
                                              seed=4000,
                                              shuffle=True,
                                              class_mode='sparse',
                                              subset='training')

valid_generator_trimmed = datagen.flow_from_directory(path_to_data,
                                              target_size=img_size,
                                              batch_size=16,
                                              seed=4000,
                                              shuffle=False,
                                              class_mode='sparse',
                                              subset='validation')
```

For training I have used a pretrained *EfficientNet* model. Top layers were not included and I have added a Dense layer. For the 30s clips dataset I have retrained this model directly without freezing any layers. With the 3s clips dataset I have expected improvement in the accuracy due to the larger dataset but the result is almost the same as with the 30s clips. Here I have first trained for couple of epochs only the added layer leaving the other layers non-trainable. Then I have unfreezed the top 150 layers leaving only *BatchNormalization* layers non-trainable. Training was done on the Google Colab using code presented below.

### 30s

```python
base_model = keras.applications.EfficientNetB0(weights='imagenet', include_top=False)
custom_model = keras.layers.GlobalAveragePooling2D()(base_model.output)
output = keras.layers.Dense(len(class_names), activation='softmax')(custom_model)
model = keras.Model(inputs=base_model.input, outputs=output)
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint('music_classifier_efficientnet_full.h5', save_best_only=True)
early_stopping = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])   
# model.summary()
```

```python
history = model.fit(train_generator_full, epochs=50, validation_data=valid_generator_full, 
                    callbacks=[early_stopping, checkpoint_cb])
```

### 3s

```python
base_model = keras.applications.EfficientNetB0(weights='imagenet', include_top=False)
for layer in base_model.layers:
    layer.trainable = False
custom_model = keras.layers.GlobalAveragePooling2D()(base_model.output)
output = keras.layers.Dense(len(class_names), activation='softmax')(custom_model)
model = keras.Model(inputs=base_model.input, outputs=output)
```

```python
early_stopping = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])   
```

```python
history = model.fit(train_generator_trimmed, epochs=50, 
                    validation_data=valid_generator_trimmed, callbacks=[early_stopping])
```

```python
weights_path = 'finetune_weights.h5'
model.save_weights(weights_path)
```

```python
NUM_UNFREEZE_LAYERS = 150

cont_model = keras.models.clone_model(model)
cont_model.load_weights(weights_path)

def unfreeze_model(model):
    for layer in model.layers[-NUM_UNFREEZE_LAYERS:]:
        if not isinstance(layer, keras.layers.BatchNormalization):
            layer.trainable = True

    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
    model.compile(
        optimizer=optimizer, loss='sparse_categorical_crossentropy', 
        metrics=['accuracy']
    )


unfreeze_model(cont_model)
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint('music_classifier_efficientnet_trimmed.h5', save_best_only=True)
```

```python
history = cont_model.fit(train_generator_trimmed, epochs=50, validation_data=valid_generator_trimmed, 
                         callbacks=[early_stopping, checkpoint_cb])
```

### Evaluation

Here we are going to compare both models using validation sets. We will also display classification reports and confusion matrices. Both models were strongly overfitting. I tried to build a CNN from scratch with stronger regularization but it didnt help.

```python
model_full = keras.models.load_model('music_classifier_efficientnet_full.h5')
model_trimmed = keras.models.load_model('music_classifier_efficientnet_trimmed.h5')
```

#### Train set

```python
model_full.evaluate(train_generator_full)
```

```python
model_trimmed.evaluate(train_generator_trimmed)
```

#### Valid set

```python
model_full.evaluate(valid_generator_full)
```

```python
model_trimmed.evaluate(valid_generator_trimmed)
```

```python
class_mapping = {i: j for j, i in valid_generator_full.class_indices.items()}
```

```python
class_mapping
```

```python
for model, generator, name in zip([model_full, model_trimmed], 
                            [valid_generator_full, valid_generator_trimmed],
                           ['full model', 'trimmed_model']):
    y_pred = model.predict(generator)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_pred_classes_names = [class_mapping[i] for i in y_pred_classes]
    
    yhat = generator.classes
    yhat_classes_names = [class_mapping[i] for i in yhat]
    
    print(name)
    print('\n')
    print(classification_report(yhat, y_pred_classes))
    plt.figure(figsize=(15,10))
    sns.heatmap(confusion_matrix(yhat, y_pred_classes), annot=True,
               xticklabels=class_mapping.values(), yticklabels=class_mapping.values())
    plt.title('Confusion matrix ' + name)
    plt.show()
```

Both models have almost the same accuracy on the validation set but model trained on trimmed clips shows smaller loss. Comparison of confusion matrices shows that the metal genre is the most correctly classified class. Also classical genre for both models shows high true positive numbers. However, the full model correctly classified most of the reggae genre while the trimmed model has a poor performance on this class. Direct comparison of both models is not possible since for the first one we only have 20 instances for each class for validation while for the second one around 200. To improve the performance one can analyze the misclassified examples in details or change the trimming method so the 3s intervals are taken in random locations (still non-overlapping). It is possible that e.g. beginnings of clips do not reflect the genre properly. The trimmed model was chosen for the deployment on Streamlit due to the smaller loss.
